# Procesamiento de imagen con Numpy

```python
import matplotlib as plt 
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt 
import urllib.request
from PIL import Image
```

## Creamos un array desde una imagen de internet

```python
url = "https://images.unsplash.com/photo-1519501025264-65ba15a82390?ixlib=rb-1.2.1&ixid=eyJhcHBfaWQiOjF9&w=1000&q=80"
internet_image = Image.open(urllib.request.urlopen(url))
```

```python
# Transformamos la imagen en arrays

image1 = np.array(internet_image)
image2 = np.array(internet_image)
image3 = np.array(internet_image)

print(image1)
```

## Explorando

```python
# Para mostrar la imagen

plt.figure(figsize=(10,20))
plt.imshow(image1)
```

```python
# Para saber la forma

# La forma es 1251 x 1000 
# el 3 representa la tercera dimension: profundidad de colores 
# Indica que la imagen esta superpuesta por tres planos
# Donde los planos son: rojo, verde y azul

# Si fuera una imagen como: (1252, 1000, 1) o (1252,1000)
# Seria de una dimension y en color blanco y negro

print(image1.shape)
1251 * 1000 * 3
```

```python
# Eligo una fila que va desde el espacio 100 y el 300
# En las columnas tomo todas las columnas ( : )
# Y por ultimo elige la profundidad que ser el rojo (0)

image2[100:300, : , 0] = 255 
image2[600:800, 100: 200, 1] = 255 

plt.figure(figsize=(10,20))
plt.imshow(image2)
```

# MASCARA CIRCULAR

```python
url = "https://st4.depositphotos.com/1035350/21608/v/600/depositphotos_216087068-stock-video-philadelphia-pennsylvania-november-2016-traffic.jpg"
internet_image = Image.open(urllib.request.urlopen(url))
```

```python
# Transformamos la imagen en arrays

image1 = np.array(internet_image)
image2 = np.array(internet_image)
image3 = np.array(internet_image)

plt.imshow(image1)
```

```python
print(image1.shape)

total_rows, total_cols, total_layers = image1.shape

X, Y = np.ogrid[:total_rows, :total_cols]

print ("X = ", X.shape, "and Y = ", Y.shape)
```

```python
# Lo usaremos para el eje x (filas)
X 
```

```python
# LO usaremos como indice para el eje y (columnas)

Y
```

```python
# Calcularemos el indice de la columna y fila central

# Obtenemos la columna y fila del medio de la imagen
center_row, center_col = total_rows / 2, total_cols / 2
print("Fila central = ", center_row, "Columna central = ", center_col)
```

```python
# Calculamos las distancias al centro y las ponemos en una matriz
# Lo elevamos al cuadrado para eliminar los valores negativos
dist_from_center = (X - center_row)**2 + (Y - center_col)**2

print(" Matriz que contiene las distancias al centro")
print("")
print(dist_from_center)
print("")
print('Distancia maxima del centro', np.amax(dist_from_center))
```

```python
# Calculamos el tamao del radio deseado
radius = (total_rows / 2)**2
print("El radio es: ", radius)
```

```python
circular_mask = (dist_from_center > radius)
print(circular_mask.shape)
print(circular_mask[300:600,600:1000])
print(circular_mask)
```

```python
# Aplicamos la mascara a la foto

image1[circular_mask] = 0
plt.figure(figsize=(10,20))
plt.imshow(image1)
```

# FILTRO DE COLOR

```python
url = "https://www.city.ac.uk/__data/assets/image/0008/408284/dubai-skyline-thumb.jpg"

internet_image = Image.open(urllib.request.urlopen(url))

image = np.array(internet_image)
plt.figure(figsize=(10,20))
print(" Imagen original")
plt.figure(figsize=(10,20))
plt.imshow(image)
```

```python
# Creamos una mscara que tomara valor true, donde existan vlores mayores a 115 verde
green_mask = image[:,:, 1] > 115 
print(" Mscara verde")
print(green_mask)
```

```python
# Operamos sobre los pixeles donde hay verdes intensos

image[green_mask, 0] = 0
image[green_mask, 1] = 0
image[green_mask, 2] = 0

print (image[green_mask])
plt.figure(figsize=(10,20))
plt.imshow(image)
```
